```python
import pandas as pd

df1 = pd.read_excel("Türkiye Düzeyinde Yoksul Kişi Sayısı Eğitim Bazlı.xls")
df2 = pd.read_excel("Türkiye Düzeyinde Yoksulluk Oranı Eğitim Bazlı.xls")
df1 = pd.DataFrame(df1)
df2 = pd.DataFrame(df2)
```

```python
kişi = df1.drop(df1.index[:3])
kişi = kişi.reset_index(drop=True)
oran = df2.drop(df2.index[:3])
oran = oran.reset_index(drop=True)
```

```python
years = list(set(kişi["Unnamed: 2"].dropna()))
years.sort()
years = [int(element) for element in years]
print(years)
```

```python
headingsPerson = list(set(kişi["Unnamed: 1"].dropna()))
headingsPerson = [element.replace("Eğitim durumu:", "") for element in headingsPerson]
print(headingsPerson)
```

```python
kişi
```

```python
holderMix = {}
for year in range(len(years)):
    temp = []
    value = years[year]
    for index in range(len(headingsPerson)):
        cell_val = kişi["Sütunlar"][year]
        temp.append(cell_val)
        year += 17
    holderMix[value] = temp
```

```python
print(holderMix[2011])
```

```python
final_data_1 = pd.DataFrame(index = pd.Index(years), columns= headingsPerson)
for year in years:
    final_data_1.loc[year] = holderMix[year]
```

```python
final_data_2 = pd.DataFrame(holderMix, index = pd.Index(headingsPerson), columns= years)
```
